```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("user_point_title.csv")
df
```

```python
book_ratingCount = (df.groupby
        (by=['title_id'])
        ['point'].mean().reset_index().rename(columns={'point': 'total_point'})
        [['title_id', 'total_point']])
```

```python
book_ratingCount
```

```python
book_ratingCount.sort_values(by=['total_point'],ascending = False).reset_index()
```
